```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("../data/raw/data.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
for replace in [" ", "-"]:
    df.columns = [column.replace(replace, "_") for column in df.columns]

#TODO DRY - Improve this by using regular expressions.
for replace in ["(", ")"]:
    df.columns = [column.replace(replace, "") for column in df.columns]
```

```python
df.columns
```

```python
df.head()
```

```python
df.tail()
```

```python
for column in ["treatment_company", "operator"]:
    df[column] = df[column].str.replace(" ", "_")
```

```python
df.head()
```

```python
df["date_on_production"] = pd.to_datetime(df["date_on_production"])
```

```python
df["year_on_production"] = df["date_on_production"].dt.strftime("%Y").astype(int)
```

```python
df.head()
```

```python
df.to_csv("../data/clean/data.csv", index=False)
```
